### 3.IF-ELIF
#### Syntax


```python
if ():
    print()
elif():
    print()
elif():
    print()
else:
    print()   
```

#### Example:


```python
age = int(input("Enter the age:"))
```

```python
if (age<18):
    print("children")
elif(age<35):
    print("Adult")
elif(age<59):
    print("Citizen")
else:
    print("Senior citizen")   
```

### 4.FOR-LOOP
#### Example


```python
lists=[23,56,46,73,23]
```

```python
for temp in lists:
    print(temp)
    
```

```python
for age in lists:
    if(age<18):
        print("children")
    elif(age<35):
        print("Adult")
    elif(age<59):
        print("Citizen")
    else:
        print("Senior citizen")   
    
    

   
```
