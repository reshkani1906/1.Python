```python
# print Odd Numbers in the list
```

```python
numbers = [20, 10, 16, 19, 25, 1, 276, 188]

for num in numbers:
    if num % 2 != 0:
        print(num, "odd")
    
```

```python
# print Even Numbers in the list
```

```python
numbers = [20, 10, 16, 19, 25, 1, 276, 188]

for num in numbers:
    if num % 2 == 0:
        print(num,"even")
```
